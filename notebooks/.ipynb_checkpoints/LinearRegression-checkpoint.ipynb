```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("homeprices.csv")
df
```

```python
plt.title("Area Vs Price")
plt.xlabel("area")
plt.ylabel("price")
plt.scatter(df.area,df.price,color="red",marker='+')
plt.show()
```

```python
area=df.drop('price',axis=1)
area
```

```python
price=df.price
price
```

```python
#equation for linear regression y=mx+c

reg=linear_model.LinearRegression()
reg.fit(area,price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
#to predict the new area 3500

new_data=pd.DataFrame([[3500]],columns=['area'])
prediction=reg.predict(new_data)
prediction=np.round(prediction,2)
prediction
```

```python
#Manual Calculation
135.78767123 * 3500 + 180616.43835616432
```

```python
p=reg.predict(df[['area']])
p=np.round(p,2)
p
```

```python
plt.title("Area Vs Price")
plt.xlabel("area")
plt.ylabel("price")
plt.scatter(df.area,df.price,color="red",marker='+')
plt.plot(df.area,reg.predict(df[['area']]),color='blue')
plt.show()
```

```python
reg.score(df[['area']],df.price)
```

```python
#Prediction for area= 6500 and 10000

new_data=pd.DataFrame([[6500]],columns=['area'])
prediction=reg.predict(new_data)
prediction=np.round(prediction,2)
prediction
```

```python
new_data=pd.DataFrame([[10000]],columns=['area'])
prediction=reg.predict(new_data)
prediction=np.round(prediction,2)
prediction
```
